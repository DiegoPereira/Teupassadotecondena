```python
import pandas as pd
import numpy as np
```

```python
df_qt_adt_con = pd.read_csv('../data/qtde_aditivos_contrato.csv', sep=';', index_col=0,dtype={'nu_CPFCNPJ': str})
df_qt_adt_cnpj = pd.read_csv('../data/qtde_aditivos_cpfcnpj.csv', sep=';', index_col=0, dtype={'nu_CPFCNPJ': str})
df_qt_con_cnpj = pd.read_csv('../data/qtde_contratos_empresa.csv', sep=';', index_col=0, dtype={'nu_CPFCNPJ': str})
df_qt_con_gan = pd.read_csv('../data/qtde_contratos_ganhos.csv', sep=';', index_col=0, dtype={'nu_CPFCNPJ': str})
df_qt_lic_tip = pd.read_csv('../data/qtde_licitacoes_por_tipo_por_cnpj.csv', sep=';', dtype={'nu_CPFCNPJ': str})
df_qt_par_tip = pd.read_csv('../data/qtde_participacoes_por_tipo_por_cnpj.csv', sep=';', dtype={'nu_CPFCNPJ': str})
df_val_con = pd.read_csv('../data/valor_contratos_por_cnpj.csv', sep=';', index_col=0, dtype={'nu_CPFCNPJ': str})
df_val_tip = pd.read_csv('../data/valor_por_licitacao_por_cnpj.csv', sep=';', dtype={'nu_CPFCNPJ': str})
```

```python
df_merge = df_qt_adt_cnpj.merge(df_qt_con_cnpj, on='nu_CPFCNPJ')
df_merge.head()
```

```python
df_merge = df_merge.merge(df_qt_con_gan, on='nu_CPFCNPJ', suffixes=('', '_ganhos'))
df_merge.head()
```

```python
df_merge = df_merge.merge(df_val_con, on='nu_CPFCNPJ')
df_merge.head()
```

```python
df_merge_two = df_qt_par_tip[['nu_CPFCNPJ', 'tp_Licitacao']].merge(df_val_tip, on=['nu_CPFCNPJ', 'tp_Licitacao'])
df_merge_two = df_merge_two[(df_merge_two.tp_Licitacao == 0) | (df_merge_two.tp_Licitacao == 2) | (df_merge_two.tp_Licitacao == 3)]
df_merge_two.head()
```

```python
df_pregao = df_merge_two[df_merge_two.tp_Licitacao == 0]
df_tomada = df_merge_two[df_merge_two.tp_Licitacao == 2]
df_convite = df_merge_two[df_merge_two.tp_Licitacao == 3]
df_pregao.head()
```

```python
df_merge_two = df_merge_two.merge(df_pregao[['nu_CPFCNPJ', 'valor_total']], on=['nu_CPFCNPJ'], suffixes=('','_pregao'))
df_merge_two.head()
```

```python
df_merge_two = df_merge_two.merge(df_convite[['nu_CPFCNPJ', 'valor_total']], on=['nu_CPFCNPJ'], suffixes=('','_convite'))
df_merge_two.head()
```

```python
df_merge_two = df_merge_two.merge(df_tomada[['nu_CPFCNPJ', 'valor_total']], on=['nu_CPFCNPJ'], suffixes=('','_tomada'))
df_merge_two.head()
```

```python
df_part_pregao = df_qt_par_tip[df_qt_par_tip.tp_Licitacao == 0][['nu_CPFCNPJ', 'participacoes']]
df_part_tomada = df_qt_par_tip[df_qt_par_tip.tp_Licitacao == 2][['nu_CPFCNPJ', 'participacoes']]
df_part_convite = df_qt_par_tip[df_qt_par_tip.tp_Licitacao == 3][['nu_CPFCNPJ', 'participacoes']]
df_part_pregao.head()
```

```python
df_merge_two = df_merge_two.merge(df_part_pregao, on='nu_CPFCNPJ')
df_merge_two = df_merge_two.merge(df_part_tomada, on='nu_CPFCNPJ',suffixes=('','_tomada'))
df_merge_two = df_merge_two.merge(df_part_convite, on='nu_CPFCNPJ',suffixes=('','_convite'))
```

```python
df_merge_two = df_merge_two.rename(columns={'participacoes': 'participacoes_pregao'})
df_merge_two.head()
```

```python
df_merge_two['participacoes_total'] = df_merge_two.participacoes_pregao + df_merge_two.participacoes_tomada + df_merge_two.participacoes_convite
df_merge_two.head()
```

```python
df_merge_two.valor_total = df_merge_two.valor_total_pregao + df_merge_two.valor_total_tomada + df_merge_two.valor_total_convite
df_merge_two.head()
```

```python
df_merge = df_merge.merge(df_merge_two, on='nu_CPFCNPJ')
df_merge.head()
```

```python
df_merge.to_csv(path_or_buf='../data/features.csv', sep=';')
```
